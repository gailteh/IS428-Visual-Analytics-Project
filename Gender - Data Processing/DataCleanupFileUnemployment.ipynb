```python
# WOMEN UNEMPLOYMENT LEVEL

import pandas as pd

# Load the data
data = pd.read_csv('Unemployment_Number_Women.csv')

# Create an empty list to store the transformed rows
transformed_rows = []

for index, row in data.iterrows():
    year = int(row['Year'])  # Convert Year to integer
    for month_num, month_name in enumerate(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'], 1):
        month_val = row[month_name]
        
        # Check if month_val is NaN, and if so, continue to the next iteration
        if pd.isna(month_val):
            continue
            
        # Construct the yearAndMonth string
        yearAndMonth = f"{year}-{month_num:02d}"
        
        # Append a dictionary to the transformed_rows list
        transformed_rows.append({
            'yearAndMonth': yearAndMonth,
            'Year': year,
            'Month': month_name,
            'Unemployment Level (in thousands)': int(month_val),
            'Gender': 'Women'
        })

# Convert the list of dictionaries to a DataFrame
transformed_data = pd.DataFrame(transformed_rows)
transformed_data = transformed_data.loc[transformed_data['yearAndMonth'] <= "2023-08"]

# Save the transformed data to a new CSV file
transformed_data.to_csv('Unemployment_Level_Women_transformed_file.csv', index=False)
```
