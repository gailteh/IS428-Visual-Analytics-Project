```python
import pandas as pd
import numpy as np
```

Bachelor degree

```python
clf = pd.read_excel("Data/Education/(Seas) Civilian Labor Force Level - Bachelor's degree and higher, 25 yrs. & over.xlsx", header=12)
year = clf.Year.values
clf = clf.loc[:,clf.columns!="Year"]

edu = pd.DataFrame()
month = list(clf.columns)
month_year_value = []
edu_att_value = []
months = {
    "Jan":[1, "January"],
    "Feb":[2, "February"],
    "Mar":[3, "March"],
    "Apr":[4, "April"],
    "May":[5, "May"],
    "Jun":[6, "June"],
    "Jul":[7, "July"],
    "Aug":[8, "Augest"],
    "Sep":[9,"September"],
    "Oct":[10, "October"],
    "Nov":[11, "November"],
    "Dec":[12, "December"]
}
month_value = []
year_value = []

for mon in month:
        for yr in year:
            if mon != "Oct" and mon != "Nov" and mon != "Dec":
                month_year_value.append(str(yr)+"-0"+str(months[mon][0]))
            else:
                month_year_value.append(str(yr)+"-"+str(months[mon][0]))
            month_value.append(months[mon][1])
            year_value.append(yr)
            edu_att_value.append("Bachelor's Degree")

edu["month_year"] = month_year_value
edu["Year"] = year_value
edu["Month"] = month_value
edu["education_attainment"] = edu_att_value

# clf_month = []
# for mon in month:
#     clf_month.extend(list(clf[mon].values))
# edu["civilian_labor_level"] = clf_month

# emp_lvl = pd.read_excel("Data/Education/(Seas) Employment Level - Bachelor's degree and higher, 25 yrs. & over.xlsx",header=12)
# emp_lvl_month = []
# for mon in month:
#     emp_lvl_month.extend(list(emp_lvl[mon].values))
# edu["employment_level"] = emp_lvl_month

# emp_pop = pd.read_excel("Data/Education/(Seas) Employment-Population Ratio - Bachelor's degree and higher, 25 yrs. & over.xlsx",header=12)
# emp_pop_month = []
# for mon in month:
#     emp_pop_month.extend(list(emp_pop[mon].values))
# edu["employment_population_ratio"] = emp_pop_month

# lab_force_part = pd.read_excel("Data/Education/(Seas) Labor Force Participation Rate - Bachelor's degree and higher, 25 yrs. & over.xlsx", header=12)
# lab_force_part_month = []
# for mon in month:
#     lab_force_part_month.extend(list(lab_force_part[mon].values))
# edu["labor_force_participation_rate"] = lab_force_part_month

unemp_lvl = pd.read_excel("Data/Education/(Seas) Unemployment Level - Bachelor's degree and higher, 25 yrs. & over.xlsx", header=12)
unemp_lvl_month = []
for mon in month:
    unemp_lvl_month.extend(list(unemp_lvl[mon].values))
edu["unemployment_level"] = unemp_lvl_month

unemp_rate = pd.read_excel("Data/Education/(Seas) Unemployment Rate - Bachelor's degree and higher, 25 yrs. & over.xlsx", header=12)
unemp_rate_month = []
for mon in month:
    unemp_rate_month.extend(list(unemp_rate[mon].values))
edu["unemployment_rate"] = unemp_rate_month

edu.tail()
```
