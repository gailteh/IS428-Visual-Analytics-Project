```python
import pandas as pd
import numpy as np
from datetime import datetime
```

##### == START == Age 20 to 24 years (level)

```python
# Read file
age_df1 = pd.read_excel("./Data/Age/(Seas) Unemployment Level - 20-24 yrs..xlsx", header=11)
age_df1.head()
```

```python
# format 
age_df1 = age_df1.melt(id_vars='Year', var_name='Month', value_name='Unemployment level')
# Create a new 'Date' column by combining 'Year' and 'Month'
age_df1['Date'] = age_df1['Year'].astype(str) + '-' + age_df1['Month']
age_df1
```

```python
#Add column for age range
age_df1["Age"] = "20 to 24 years"
age_df1
```

```python
#Convert date into ie. 1969-01
age_df1["Date"] = pd.to_datetime(age_df1["Date"], format = "%Y-%b").dt.strftime("%Y-%m")
age_df1
```

```python
#define cutoff date which is Aug 2023
age_df1 = age_df1.loc[age_df1['Date'] <= "2023-08"]
age_df1
```
